# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Initializing the independent variable, x

```python
x = np.linspace(-10, 10, 100)
```

#1- Linear Units

```python
y=x
plt.plot(x, y, '-b')
plt.title('Linear Activation')
plt.xlabel('Input Value')
plt.ylabel('Activation Output')
plt.show()
```

#2- Sigmoid Function

```python
y = 1/(1 + np.exp(-x))
  
plt.plot(x, y)
plt.xlabel("Input Value")
plt.ylabel("Activation Output")
plt.title("Sigmoid Activation")
plt.show()
```

#3- Hyperbolic Tangent Units

```python
y=np.tanh(x)
plt.plot(x, y,'-b')
plt.title('Hyperbolic Tangent Activation')
plt.xlabel('Input Value')
plt.ylabel('Activation Output')
plt.show()
```

# 4- Rectified Linear Units

```python
def ReLU(x):
   return (np.maximum(x,0))
```

```python
y = ReLU(x)
plt.plot(x, y,'-b')
plt.title('ReLU Activation')
plt.xlabel('Input Value')
plt.ylabel('Activation Output')
plt.show()
```
